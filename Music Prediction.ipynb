```
songs_schema = StructType([
    StructField("song_id", StringType(), True),
    StructField("artist_id", StringType(), True),
    StructField("publish_time", IntegerType(), True),
    StructField("songs_init_plays", IntegerType(), True),
    StructField("language", IntegerType(), True),
    StructField("gender", IntegerType(), True)])
users_schema = StructType([
    StructField("user_id", StringType(), True),
    StructField("song_id", StringType(), True),
    StructField("gmt_create", IntegerType(), True),
    StructField("action_type", IntegerType(), True),
    StructField("ds", IntegerType(), True)])

songs_path = "/FileStore/tables/mars_tianchi_songs.csv"
users_path = "/FileStore/tables/mars_tianchi_user_actions.csv"

sdf_songs = spark.read.format('csv') \
    .option("header", False) \
    .option("dateFormat", "yyyyMMdd") \
    .option("inferSchema", "true") \
    .option("nullValue", "null") \
    .schema(songs_schema) \
    .load(songs_path)
sdf_users = spark.read.format('csv') \
    .option("header", False) \
    .option("dateFormat", "yyyyMMdd") \
    .option("inferSchema", "true") \
    .option("nullValue", "null") \
    .schema(users_schema) \
    .load(users_path)

sdf_songs.show()
sdf_users.show()
```

# K-Means Clustering

```
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator

from pyspark.ml.feature import PCA as PCAml
```

```
assemble=VectorAssembler(inputCols=[
'action_type',
'songs_init_plays',
'language',
'gender',
'play_time',
'duration'],outputCol = 'music_features')

assembled_data=assemble.transform(df_final)
```

```
assembled_data.show()
```

```
silhouette_scores=[]
evaluator = ClusteringEvaluator(predictionCol='prediction', \
featuresCol='music_features', \
metricName='silhouette', distanceMeasure='squaredEuclidean')

for K in range(2,10):
    KMeans_algo=KMeans(featuresCol='music_features', k=K)
    KMeans_fit=KMeans_algo.fit(assembled_data)
    KMeans_transform=KMeans_fit.transform(assembled_data)

    evaluation_score=evaluator.evaluate(KMeans_transform)
    silhouette_scores.append(evaluation_score)
```

这里跑了45mins，原因可能是表太大，且涉及太多特征

```
fig, ax = plt.subplots(1,1, figsize =(10,8))
ax.plot(range(2,10),silhouette_scores)
ax.set_xlabel('Number of Clusters')
ax.set_ylabel('Silhouette Score')
```

根据轮廓分数图silhouette选取k=4

```
KMeans_algo=KMeans(featuresCol='music_features', k=4) 
KMeans_Model=KMeans_algo.fit(assembled_data)
KMeans_Assignments=KMeans_Model.transform(assembled_data)
```

## PCA visualization

```
pca = PCAml(k=2, inputCol="music_features", outputCol="pca")
pca_model = pca.fit(assembled_data)
pca_transformed = pca_model.transform(assembled_data)
```

```
# extract the principal components
x_pca = np.array(pca_transformed.rdd.map(lambda row: row.pca).collect())
```

```
# retrieve the cluster assignments from k-means assignments
cluster_assignment = np.array(KMeans_Assignments.rdd.map(lambda row: row.prediction).collect()).reshape(-1,1)
```

```
pca_data = np.hstack((x_pca,cluster_assignment))

#FacetGrid
pca_df = pd.DataFrame(data=pca_data, columns=("1st_principal", "2nd_principal","cluster_assignment"))
sns.FacetGrid(pca_df,hue="cluster_assignment", height=6).map(plt.scatter, '1st_principal', '2nd_principal' ).add_legend()

plt.show()
```

```
artist_part = plays_sdf.repartition(spark.sparkContext.defaultParallelism, ['artist']).cache()
artist_part.explain()
```

```
# note: dataType  should be an instance of, remember the ()
result_schema = StructType([
                  StructField('ds', TimestampType()),\
                  StructField('artist', StringType()),\
                  StructField('y', DoubleType()),\
                  StructField('yhat', DoubleType()),\
                  StructField('yhat_upper', DoubleType()),\
                  StructField('yhat_lower', DoubleType())
])
```

引入Pandas UDFs(a.k.a. Vectorized UDFs)
建立在Apache Arrow（用于直接在JVM和Python驱动程序/执行器之间交换数据，序列化成本接近于零（反）序列化）之上，允许 矢量化 操作，低开销高性能  
GROUPED_MAP针对split-apply-combine这种分析模式，对每一组应用这个操作

```
from pyspark.sql.functions import pandas_udf, PandasUDFType

@pandas_udf(result_schema, PandasUDFType.GROUPED_MAP)
def forecast_sales(music_pd):
  model = Prophet(interval_width=0.95, seasonality_mode= 'multiplicative', weekly_seasonality=True, yearly_seasonality=True)
  model.fit(music_pd)
  future_pd = model.make_future_dataframe(periods=5, freq='w')
  forecast_pd = model.predict(future_pd)
  f_pd = forecast_pd[['ds', 'yhat', 'yhat_upper', 'yhat_lower']].set_index('ds')
  st_pd = music_pd[['ds', 'artist', 'y']].set_index('ds')
  result_pd = f_pd.join(st_pd, how='left')
  result_pd.reset_index(level=0, inplace=True)
  result_pd['artist'] = music_pd['artist'].iloc[0]
  return result_pd[['ds', 'artist', 'y', 'yhat', 'yhat_upper', 'yhat_lower']]
```

## Model Ⅰ
置信区间0.95，乘法季节性，预测60天（两个月）更多配置参见官网

```
from pyspark.sql.functions import current_date
results = artist_part.groupby('artist').apply(forecast_sales).withColumn('training_date', current_date())
results.cache()
results.show()
```

```
#split our data
split_date = '2015-07-30'
data_train = original_df.loc[original_df.index <= split_date].copy()
data_test = original_df.loc[original_df.index > split_date].copy()
```

```
def create_features(df, label=None):
    df['date'] = df.index
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.weekofyear
    X = df[['hour','dayofweek','quarter','month','year',
           'dayofyear','dayofmonth','weekofyear']]
    if label:
        y = df[label]
        return X, y
    return X
```

```
X_train, y_train = create_features(data_train, label='plays')
X_test, y_test = create_features(data_test, label='plays')
X_train
```

```
import xgboost as xgb
from xgboost import plot_importance, plot_tree
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```
reg = xgb.XGBRegressor(n_estimators=1000)
```

```
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        early_stopping_rounds=50,
       verbose=False)
```

```
_ = plot_importance(reg, height=0.9)
```
